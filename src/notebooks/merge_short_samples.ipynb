```python
import os
import librosa
import glob
import pandas as pd
```

```python
ANNOTATIONS_FILE = glob.glob("/Users/test/Documents/Projects/Master/nips4bplus/raw_audio/annotations/*csv")
DESTINATION_FOLDER = "/Users/test/Documents/Projects/Master/nips4bplus/raw_audio/new_annotations"
MIN_SIZE = 0.5
MAX_SIZE = 2.0
len(ANNOTATIONS_FILE)
```

```python
for annotation_file in ANNOTATIONS_FILE:
    df = pd.read_csv(annotation_file, names=["start", "duration", "classname"])
    print(annotation_file)
    outfile = os.path.join(DESTINATION_FOLDER, os.path.basename(annotation_file))
    os.makedirs(os.path.dirname(outfile), exist_ok=True)
    with open(outfile, 'w') as fs:
        for group_name, group in df.groupby("classname"):
            rows = group.values
            class_name  = group_name
            # print("All", rows)
            start_ = rows[0][0]
            end_ = rows[0][1]
            for i in range(len(rows)-1):
                curr_ = rows[i]
                next_ = rows[i+1]
                diff = next_[0] - (curr_[0]+curr_[1])
                sample_size = next_[0]+next_[1]-start_
                if diff < MIN_SIZE and  sample_size < MAX_SIZE:
                    end_ = max(end_, (next_[0]+next_[1])-start_)
                else:
                    # print("fin",start_, end_, group_name)
                    fs.write(f"{start_},{end_},{group_name}\n")
                    start_ = next_[0]
                    end_ = next_[1]
            # print("last", start_, end_)
            fs.write(f"{start_},{end_},{group_name}\n")
            # print()
            # break
        # break
        pass
```
